```python
# fit tracks within a specific time interval
phi_deg = 90. # spin is down
# theta_deg is camera angle of 45 from above
# t0_frame is impact frace from above
# diam is marble diameter in pixels from above
# fps is frames per second from above

# fitting spin prior to impact
ntl = 20 # min number of points in each track
df = 55  # how many frames in interval
f0 = 10   # start frame
i = 0  # where to store information in arrays!
fmax = f0+df # end frame of interval 
prefix = oroot + '_fit_'
ofile = mkostring(prefix,'.png',i)  # outfile
tmid,dt,omega,de_omega=omega_df(t2b,diam,fps,t0_frame,cx0,cy0,\
            f0,df,ntl,theta_cam_deg,phi_deg,ofile)
print('t {:.4f} s pm {:.4f} omega={:.1f} s^-1 pm {:.1f}'.format(\
                tmid,dt,omega,de_omega))
t_spin_arr[i] = tmid  # store fit data
dt_spin_arr[i] = dt
om_arr[i] = omega
de_om_arr[i] = de_omega
istore = i  # where we just stored info into the arrays

fmax = f0+df # end frame of interval just done 
print (istore,f0,fmax,df)
```

```python
ntl = 15  # min points in a track
df=20 # how many frames used for each interval
# doing fits for 10 more intervals, starting from previous in array
# starting near impact
# fitting spin for 10 more short intervals during impact
df0=df/2   # shift in first frame number for each spin measurement
for i in range(10):  # specify  here how many spin measurements to do
    f0 = fmax - 5 + i*df0   # start frame near impact
    j = istore+1+i  # where to store data
    ofile = mkostring(prefix,'.png',j)  # outfile
    tmid,dt,omega,de_omega=omega_df(t2b,diam,fps,t0_frame,cx0,cy0,\
                f0,df,ntl,theta_cam_deg,phi_deg,ofile)
    t_spin_arr[j] = tmid  # store fit data
    dt_spin_arr[j] = dt
    om_arr[j] = omega
    de_om_arr[j] = de_omega
    
fmax =  f0 + df  # end frame of last interval
istore = j # where we just stored data
print (istore,f0,fmax,df)
```

```python
# return  acceleration vectors
# that have more smoothing past time tdiv
def asmo(tarr_even,xarr_smo,yarr_smo,ax_smo,ay_smo,alen,tdiv):
    ax_smo2 = np.copy(ax_smo)
    ay_smo2 = np.copy(ay_smo)
    dt = tarr_even[1]-tarr_even[0]  # sampling rate
    ax_smo2 = savgol_filter(xarr_smo, alen, 2,mode='nearest',deriv=2)/dt**2
    ay_smo2 = savgol_filter(yarr_smo, alen, 2,mode='nearest',deriv=2)/dt**2
    jjj = (tarr_even <= tdiv)
    ax_smo2[jjj] = ax_smo[jjj]  # prior to tdvi is the same
    ay_smo2[jjj] = ay_smo[jjj]
    return ax_smo2,ay_smo2
    
```

```python
# make 7 plots of the cm trajectory and spin!
def plt7(tarr,xarr,yarr,tarr_even,xarr_smo,yarr_smo,vx,vx_smo,vy,vy_smo,\
         ax,ax_smo,ay,ay_smo,\
         t_spin_arr,dt_spin_arr,om_arr,dom_arr,\
         ofile):

    tline = np.array([-0.2,np.max(tarr_even)])  #some horizontal lines
    zline = tline*0
    fs=18  #fontsize
    nwin=7
    fig,axarr = plt.subplots(nwin,1,figsize=(7,11),sharex=True)
    plt.subplots_adjust(left=0.20, right=0.97, top=0.99, bottom=0.15, \
        wspace=0.22, hspace=0.0)
    for i in range(nwin):
        axarr[i].tick_params(axis='both', which='major', labelsize=16)
        axarr[i].tick_params(axis='both', which='minor', labelsize=14)
    
    gunits =  9.8e2  #cm^2/s
    
    # horizontal lines
    for i in range(nwin):  
        axarr[i].plot(tline,zline,'-',color='gray',lw=0.5)
    
    # points and lines
    axarr[0].plot(tarr,xarr,'r.')
    axarr[1].plot(tarr,yarr,'g.')
    axarr[0].plot(tarr_even,xarr_smo,'k-')
    axarr[1].plot(tarr_even,yarr_smo,'k-')
    axarr[2].plot(tarr_even,vx,'b.')
    axarr[2].plot(tarr_even,vx_smo,'k-')
    axarr[3].plot(tarr_even,vy,'m.')
    axarr[3].plot(tarr_even,vy_smo,'k-')
    axarr[4].plot(tarr_even,ax/gunits,'c.')
    axarr[5].plot(tarr_even,ay/gunits,'.',color='orange')
    
    # plot spins
    rad_cm = (diam/2)*pixscale/10  # marble radius in cm
    axarr[6].errorbar(t_spin_arr,om_arr,\
            xerr=dt_spin_arr,yerr=dom_arr,fmt='.',color='k')
    axarr[6].plot(t_spin_arr,om_arr,'rs',label=r'$\omega$')
    axarr[6].plot(tarr_even[6:],vx_smo[6:]/rad_cm,'-',color='gray',label=r'$v_x/R$')  
    
    # use different smoothing before and after this time
    tdiv = 0.021; alen=51
    ax_smo2, ay_smo2= asmo(tarr_even,xarr_smo,yarr_smo,ax_smo,ay_smo,alen,tdiv)
    axarr[4].plot(tarr_even,ax_smo2/gunits,'k-')
    axarr[5].plot(tarr_even,ay_smo2/gunits,'k-')
    
    #labels
    axarr[nwin-1].set_xlabel(r't (s)',fontsize=fs)
    axarr[0].set_ylabel('x (cm)',fontsize=fs)
    axarr[1].set_ylabel('z (cm)',fontsize=fs)
    axarr[2].set_ylabel(r'$v_x$ (cm/s)',fontsize=fs)
    axarr[3].set_ylabel(r'$v_z$ (cm/s)',fontsize=fs)
    axarr[4].set_ylabel(r'$a_x/g$',fontsize=fs)
    axarr[5].set_ylabel(r'$a_z/g$',fontsize=fs)
    axarr[6].set_ylabel(r'$\omega$ (1/s)',fontsize=fs)
    
    #hard limits
    axarr[0].set_xlim([-0.020,max(tarr_even)])
    axarr[4].set_ylim(top=5);  # some acceleration limits
    axarr[5].set_ylim(bottom=-3)
    axarr[1].set_ylim(top=0.9)
    axarr[6].set_ylim(top=160)
    
    #draw some vertical lines
    for i in range(nwin):  
        ylim = axarr[i].get_ylim()
        vline = ylim
        t0pos = [0.0,0.0]
        axarr[i].plot(t0pos,vline,'k-')
    # save a figure
    if (len(ofile)> 3):
        plt.savefig(ofile)
   
    
```
